```python
import requests
import os.path

xml_url = "https://raw.githubusercontent.com/andrewssobral/vehicle_detection_haarcascades/master/cars.xml"
image_url = "https://cdn.kaktusgame.eu/LSAI/cars.jpg"

if os.path.exists('./cars.xml') == False:
    xml_response = requests.get(xml_url)
    with open('./cars.xml', 'wb') as f:
        f.write(xml_response.content)
if os.path.exists('./cars.jpg') == False:
    image_response = requests.get(image_url)
    with open('./cars.jpg', 'wb') as f:
        f.write(image_response.content)
```
